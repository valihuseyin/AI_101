Mean Squared Error (MSE): Modelin tahminlerinin gerçek değerlerden ne kadar uzak olduğunu ölçer. Daha düşük değerler, daha iyi bir model anlamına gelir.
R^2 Score: Modelin verideki varyansı ne kadar iyi açıkladığını gösterir. 1'e yakın değerler, modelin iyi olduğunu gösterir.
Bu adımları takip ederek tek özellikli veri setinizle bir lineer regresyon modeli geliştirip değerlendirelim.

```python
#Veri setimiz : Resklam bilgilerini veren bir setimiz var. Burada sadece TV reklamları ve satış sonuçları var.
# hedef y : Sales satışlarımız.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
LRM = LinearRegression()
```

```python
#read data
df = pd.read_csv('reklam.csv')
```

```python
df
```

```python
x =df [['TV']]
y =df[['Sales']]
```

```python
x
```

```python
y
```

```python
#train and test data
# Eğitim ve test setlerine ayırma
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# modelimizi oluştur  ve modeli besleyelim
LRM.fit(x_train,y_train)
```

```python
# Tahmin yapma
tahmin = LRM.predict(x_test)
```

```python
tahmin
```

```python
# Değerlendirme metrikleri
mse = mean_squared_error(y_test, tahmin)
r2 = r2_score(y_test, tahmin)
```

# Model Değerlendirmesi
## Mean Squared Error (MSE): Modelin tahminlerinin gerçek değerlerden ne kadar uzak olduğunu ölçer. Daha düşük değerler, daha iyi bir model anlamına gelir.
## R^2 Score: Modelin verideki varyansı ne kadar iyi açıkladığını gösterir. 1'e yakın değerler, modelin iyi olduğunu gösterir.

```python
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
# Sonuçların görselleştirilmesi
plt.scatter(x, y, color='blue', label='Gerçek Veriler')
plt.plot(x_test, tahmin, color='red', linewidth=2, label='Tahmin Edilen Doğru')
plt.title('Lineer Regresyon Modeli')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(len(y_test)), y_test.values, label="Gerçek", marker='o')
plt.plot(range(len(tahmin)), tahmin, label="Tahmin", marker='x')
plt.xlabel("Örnek Index")
plt.ylabel("Satış")
plt.title("Gerçek vs Tahmin Satış Değerleri")
plt.legend()
plt.show()
```
