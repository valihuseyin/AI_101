```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.figure(figsize=(10, 7))
```

```python
plt.title('Satışlar') # başlık ekleme
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"r-.") # data kaynağı verme x,y, ve dottype
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"-") # data kaynağı verme x,y, ve dottype
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"y*") # data kaynağı verme x,y, ve dottype :y* yellow yıldız demek
plt.show()
```

```python
t = np.arange(0, 2.5, 0.1) # 0 ile 2.5 arası ve 0.1 artış ile array oluşturur.
t
```

```python
y1 = np.sin(np.pi*t) # x değerlerine göre sin hesapla, sin dalgası oluşturma
```

```python
plt.plot(t, y1)
plt.show()
```

```python
# birden fazla sin dalgası oluşturalım. ve aynı anda aynı grafik üzerinde gösterelim.
y2 = np.sin(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
```

```python
 plt.plot(t, y1, 'b*', t, y2, 'g^', t, y3, 'ys')
#blue yıldız,
plt.show()
```

```python
plt.plot(t, y1, 'b--', t, y2, 'g', t, y3, 'r-.')
plt.show()
```

```python
#subplot gösterim
t = np.arange(0, 5, 0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)


plt.subplot(211)
 #211-->2: satır sayısı, 1: stün sayısı 1: devam eden grafiğin hangi stünda olduğu
plt.plot(t, y1, 'b-.')
plt.subplot(212)
plt.plot(t, y2, 'r--.')
```
